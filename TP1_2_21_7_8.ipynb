* __En la columna "firstChar" tendremos al primer caracter de los titulos__

```python
mediana_caracter=antiguedad_apenom
mediana_caracter['firstChar']=mediana_caracter['title'].str.get(0)
mediana_caracter.head()
```

```python
print(mediana_caracter.shape)
```

```python
mediana_caracter.firstChar.value_counts()
```

* __Ahora creamos una columna donde se obtenga la diferencia de dias que existe entre la ultima revisión y el dia de hoy así tendremos un promedio de dias o de años.__

```python
mediana_caracter['difDays']=date_diff(datetime.datetime.today(),mediana_caracter.loc[:,'revision_timestamp'])
mediana_caracter.head()
```

```python
final=mediana_caracter[['firstChar','difDays']]
```

```python
final.head()
```

```python
final=final.groupby(['firstChar']).agg({'difDays':['mean','count','sum']})
```

```python
final.reset_index(inplace=True)
final.head()
```

```python
final.columns
```

```python
final.columns=['firstChar','mean_difDays','count_difDays','sum_difDays']
final.head()
```

* __Como podemos observar tenemos el promedio de diferencia de días para cada caracter asi como su cantidad y com la suma. Ademas mostraremos la antiguedad en años__

```python
final['mean_anios']=(final['mean_difDays']/365.2425).astype(np.int16)
final.head(20)
```

* __En lo siguiente se muestra los caracteres que tuvieron mayor aparición como el primer carácter de title.__

```python
final.nlargest(50,'count_difDays')
```
