 # Ejercicio 56
 #### Calcule la cantidad de acciones realizadas por usuarios según día de la semana.
***
***

```python
logs_56=logs[['timestamp','action']]
logs_56.head()
```

```python
logs_56.dropna(subset=['action'], inplace=True)
logs_56.info()
```

```python
logs_56['timestamp'] = pd.to_datetime(logs_56['timestamp'])
logs_56['weekday'] = logs_56['timestamp'].dt.day_name()
logs_56.head()
```

```python
logs_56=logs_56[['action','weekday']]
logs_56['cantidad']=1
logs_56.head()
```

```python
logs_56=pd.pivot_table(logs_56,values='cantidad',columns='weekday',index='action',aggfunc=np.sum)
logs_56.fillna(0,inplace=True)
logs_56.head()
```

```python
logs.action.unique()
```

```python
logs_58=logs[['comment','action','title']]
logs_58.head()
```

```python
logs_58=logs_58.loc[logs_58['action']=='block']
logs_58.info()
```

```python
logs_58.dropna(subset=['comment','title'], inplace=True)
logs_58.info()
```

```python
logs_58.head(10)
```

```python
new = logs_58["title"].str.split(":", n = 1, expand = True)
#new[0]='usuario...
logs_58["users_id_ip"]= new[1]

logs_58.drop(columns =["title"], inplace = True)
logs_58.head()
```

```python
logs_58.users_id_ip.value_counts()
```
